## Problem 5
Use the Pokemon dataset to create a clustermap with color. First, filter the dataframe to only keep data with `Type 1` equals to following values: `Water`, `Normal`, `Grass`, `Bug` and `Psychic`. Annotate the dendrogram using different colors for these five different `Type 1` values. Use `Name` as the index.

pokemon_data is in https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv

```python
# Problem 5
# Your code here . . .
pokemon_csv = "https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv"
pokemon = pd.read_csv(pokemon_csv)
pokemon.info()
pokemon
```

```python
desired_types = ['Water', 'Normal', 'Grass', 'Bug', 'Psychic']
filtered_pokemon = pokemon[pokemon['Type 1'].isin(desired_types)]
filtered_pokemon.set_index('Name', inplace=True)
colors = {"Water": "blue", "Normal": "grey", "Grass": "yellow", "Bug": "orange", "Psychic": "purple"}
pokemon_color = filtered_pokemon['Type 1'].map(colors)
filtered_pokemon
```

```python
filtered_pokemon = filtered_pokemon[["Total", "HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]]

g = sns.clustermap(data=filtered_pokemon,
               row_colors=pokemon_color,
               figsize=(18, 15),
               cmap='viridis')
g.fig.suptitle("Pokemon Clustering", fontsize=15, y=1.02)
for label, color in colors.items():
    g.ax_row_dendrogram.bar(0, 0, color=color, label=label, linewidth=0)

g.ax_row_dendrogram.legend(
    title="Type",
    loc="center",
    bbox_to_anchor=(0.2, 0.5)
)
plt.show()
```
